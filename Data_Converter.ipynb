```python
import os
import pandas as pd
import numpy as np
import io
from IPython.display import clear_output
```

```python
path = "C:\\Users\\wolff\\Desktop\\AI for Climate\\Datasets\\Daily mean sea level pressure"

log_interval = 100
files = len(os.listdir(path))
list_of_dates = []

def get_data(path):
    zaehler = 0
    counter = 0
    frame = pd.DataFrame()
    for filename in os.listdir(path):
        counter += 1
        if "STAID" in filename:
            content = open(path + "\\" + filename,'r', encoding="utf-8").readlines()
            if "FRANCE" or "GERMANY" or "BELGIUM" in content[16]:
                zaehler = zaehler + 1
                if zaehler == 20:
                    data = content[20:len(content)]
                    data = [line.replace(" ", "").replace("\n", "").split(",") for line in data]
                    df = pd.DataFrame(data[1:], columns=data[0])
                    if int(df["DATE"][0]) < 19800101:
                        save = [value[3] for value in df.values.tolist() if int(value[2]) > 19800101]
                        if len(save) == 14944:
                            frame[filename] = save
                    zaehler = 0
        if counter % log_interval == 0:
            clear_output(wait=True)
            print("You reached ", (counter / files)*100, "percent")
    return frame
        

data = get_data(path)
print(data)
```
